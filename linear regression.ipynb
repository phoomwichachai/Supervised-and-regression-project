```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
#ex1
data = pd.read_csv('Salary_Data.csv')
x = data['YearsExperience'].to_numpy().reshape(-1,1)
y = data['Salary'].to_numpy().reshape(-1,1)

print(data.info(), data.describe())
n = x.size
```

```python
#1.2
xsum = x.sum()
ysum = y.sum()
x_sum = (x**2).sum()
xy = (x*y).sum()
x_mean = x.mean()**2
ymean = y.mean()
xmean = x.mean()
upper_eq = ((xy)-((xsum)*(ysum))/n)
lower_eq = (x_sum) - (n*x_mean)
m = upper_eq/lower_eq
print(m)
b = ymean - (m*xmean)
print(b)
y_pre = m * x + b
print("Mean squared error1: %.7f" % mean_squared_error( y, y_pre))
#1.3
plt.plot(x ,y_pre ,"r-")
plt.scatter(x,y)
plt.show()
```

```python
#ex3
#3.1-3.2
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
data = pd.read_csv('Salary_Data.csv')
x = np.array(data['YearsExperience'])
y = np.array(data['Salary'])
X = x.reshape(-1,1)
Y = y.reshape(-1,1)
model = LinearRegression()
model.fit(X, Y)
print(model.intercept_ , model.coef_)
y_pre = model.predict(X)
print("Mean squared error: %.3f" % mean_squared_error( y_pre, Y))
plt.plot(X, y_pre ,"r-")
plt.scatter (X,Y)
```

```python
#ex4
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
data = pd.read_csv('Sales.csv')
#4.2
print(data.info(), data.describe())
X = np.array(data['Newspaper']).reshape(-1,1)
Y = np.array(data['Sales']).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.7 ,test_size = 0.3, random_state=2)
model = LinearRegression()
model.fit(X_train, y_train)
print(model.intercept_ , model.coef_)
y_pre = model.predict(X_test)
print("Mean squared error: %.3f" % mean_squared_error(y_test, y_pre))
print('r_square_value :',r2_score(y_test, y_pre))
plt.scatter(X,Y)
plt.plot(X_test, y_pre  ,"r-")
```
